```python
img = Image.open('testbill.JPEG')
plt.imshow(img)
```

```python
enhancer= ImageEnhance.Contrast(img)
contrastim= enhancer.enhance(1.5)
blim=contrastim.filter(ImageFilter.BoxBlur(1))
enhancer = ImageEnhance.Sharpness(contrastim)
sharpenim=enhancer.enhance(1.5)
plt.imshow(sharpenim)
```
